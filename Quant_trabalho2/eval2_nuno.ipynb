```python
# Verificação usando Qiskit
qc = QuantumCircuit(1)
qc.x(0)
qc.z(0)
qc.h(0)

unitary = Operator(qc).data
print("Unitary matrix:")
print(unitary)
```

### 2.3 Exercícios simples 3

1. Crie um circuito quântico que produza o estado de Bell: $\frac{1}{\sqrt{2}}(|01\rangle + |10\rangle)$. Utilize o Qiskit para construir o circuito e statevector para verificar o seu resultado.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Create a quantum circuit for the Bell state |psi⟩ = 1/sqrt(2) (|01⟩ + |10⟩)
expected_bell_state = Statevector([0, 1, 1, 0]) / np.sqrt(2)

qc = QuantumCircuit(2)
qc.x(1)
qc.h(0)
qc.cx(0, 1)

print(qc)

# Get the statevector of the circuit
statevector = Statevector.from_instruction(qc)
print("Statevector of the circuit:", statevector)

# Verify if the statevector matches the expected Bell state
if statevector.equiv(expected_bell_state):
    print("The circuit produces the expected bell state.")
else:
    print("The circuit does not produce the expected state.")
```
